```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sent_tokens=nltk.word_tokenize(s)
sent_tokens1=nltk.word_tokenize(s1)
print(sent_tokens1)
def Lem(text):
  return nltk.word_tokenize(text.lower())
def Lem1(text):
  return nltk.word_tokenize(text.lower())
```

```python
def response(user):
  user = user.lower() 
  robo_response = ''
  sent_tokens.append(user)
  TfidfVec = TfidfVectorizer(tokenizer = Lem, stop_words='english')
  tfidf = TfidfVec.fit_transform(sent_tokens)
  vals = cosine_similarity(tfidf[-1], tfidf)
  idx = vals.argsort()[0][-2]
  flat = vals.flatten()
  flat.sort()
  score = flat[-2]
  robo_response1 = ''
  sent_tokens1.append(user)
  TfidfVec1 = TfidfVectorizer(tokenizer = Lem1, stop_words='english')
  tfidf1 = TfidfVec.fit_transform(sent_tokens1)
  vals1 = cosine_similarity(tfidf1[-1], tfidf1)
  idx1 = vals1.argsort()[0][-2]
  flat1 = vals1.flatten()
  flat1.sort()
  score1 = flat1[-2]
  print(score,score1)
  if (score>score1):
    response = 'negative'
  elif score<score1:
    response = 'postive'
  else:
    response='moderate'
  try:
      sent_tokens.remove(user)
  except:
    pass
  try:
        sent_tokens1.remove(user)
  except:
    pass
  return response
```
